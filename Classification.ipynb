```python
spheres = data.sphere.unique()
```

```python
df_res = pd.DataFrame()

for sphere in spheres:
    df_sphere = data[data['sphere'] == sphere][:]
    df_res = df_res.append(df_sphere, ignore_index=True)
```

```python
from nltk.corpus import stopwords
from pymystem3 import Mystem
from nltk.stem.snowball import SnowballStemmer
mystem = Mystem()
```

```python
stemmer = SnowballStemmer("russian")
russian_stopwords = stopwords.words("russian")
russian_stopwords.extend(['…', '«', '»', '...', 'Здравствуйте', 'здравствуйте', 'т.д.', 'т', 'д'])
from nltk import word_tokenize

stemmed_texts_list = []
for text in df_res['text_prep']:
    tokens = word_tokenize(text)
    stemmed_tokens = [stemmer.stem(token) for token in tokens if token not in russian_stopwords]
    text = " ".join(stemmed_tokens)
    stemmed_texts_list.append(text)

df_res['text_stem'] = stemmed_texts_list
```
